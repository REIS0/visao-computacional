```
import numpy as np
import matplotlib.pyplot as plt
```

```
from PIL import Image

im = Image.open("data/escaladecinza.jpg")
im.show()
```

```
im_array = np.array(im)
im_array.shape
```

```
im_array
```

```
hist = np.zeros(256)

for y in range(im_array.shape[0]):
    for x in range(im_array.shape[1]):
        value = im_array[y][x][0]
        hist[value] += 1

plt.figure(figsize=(15, 6))
plt.bar(np.arange(256), hist)
plt.show()
```

```
im_bin = np.empty(im_array.shape, dtype=np.uint8) # precisa colocar o tipo como uint 8 bits

for y in range(im_array.shape[0]):
    for x in range(im_array.shape[1]):
        if im_array[y][x][0] > 50:
            im_bin[y][x][0] = 255
            im_bin[y][x][1] = 255
            im_bin[y][x][2] = 255
        else:
            im_bin[y][x][0] = 0
            im_bin[y][x][1] = 0
            im_bin[y][x][2] = 0

Image.fromarray(im_bin).show()
```

```
im_bin = np.empty(im_array.shape, dtype=np.uint8) # precisa colocar o tipo como uint 8 bits

for y in range(im_array.shape[0]):
    for x in range(im_array.shape[1]):
        if 100 < im_array[y][x][0] < 180:
            im_bin[y][x][0] = 0
            im_bin[y][x][1] = 0
            im_bin[y][x][2] = 0
        else:
            im_bin[y][x][0] = 255
            im_bin[y][x][1] = 255
            im_bin[y][x][2] = 255

Image.fromarray(im_bin).show()
```
